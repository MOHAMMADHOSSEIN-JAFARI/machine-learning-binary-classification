```python
import os
from os import listdir
import pandas as pd
# Basic packages
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random
import pickle

# Image preprocessing
import pathlib
import PIL
import PIL.Image
import cv2
```

```python
dataset = []
cat_address = '/kaggle/input/pured-cats-dogs/CatsDogs/Cats'
dog_address = '/kaggle/input/pured-cats-dogs/CatsDogs/Dogs'
```

```python
def cat_generator(): 
    for item in os.listdir(cat_address):
        try:
            tensor = cv2.imread(os.path.join(cat_address,item), cv2.IMREAD_COLOR)
            new_tensor= cv2.resize(tensor, (64,64))
            dataset.append([new_tensor, 0])
        except: 
            pass
    
    
    
```

```python
cat_generator()
```

```python
len(dataset)
```

```python
def dog_generator(): 
        for item in os.listdir(dog_address):
            try: 
                tensor = cv2.imread(os.path.join(dog_address,item), cv2.IMREAD_COLOR)
                new_tensor= cv2.resize(tensor, (64,64))
                dataset.append([new_tensor, 1])
            except: 
                pass
```

```python
dog_generator()
```

```python
random.shuffle(dataset)
```

```python
x_data = []
y_data = []
```

```python
for array, label in dataset:
    x_data.append(array)
    y_data.append(label)
    
```

```python
x_data = np.array(x_data).reshape(-1, 64, 64, 3)
```

```python
x_data.shape
```

```python
import pickle
pickle_out = open('/kaggle/working/x_data.pickle','wb')
pickle.dump(x_data, pickle_out)
pickle_out.close()

pickle_out = open('/kaggle/working//y_data.pickle','wb')
pickle.dump(y_data, pickle_out)
pickle_out.close()
```
