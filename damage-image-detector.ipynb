
**The script below finds damaged images in both cats and dogs folder. These damaged pictures will  not allow the models to work correctly. 
In the Cats folder, the image number 666 and in the dogs folder the image number 11702 are deleted manually from folders.**

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import os
from os import listdir
from PIL import Image
```

```python
# Finding broken or damaged images in cats folder     
  
for filename in listdir('/kaggle/input/cat-dog-dataset/CatsDogs/Cats'):
  if filename.endswith('.jpg'):
    try:
      img = Image.open('/kaggle/input/cat-dog-dataset/CatsDogs/Cats/'+filename) # open the image file
      img.verify() # verify that it is, in fact an image
    except (IOError, SyntaxError) as e:
      print('Damaged Image:', filename)
```

```python
# Finding broken or damaged images in dogs folder    
for filename in listdir('/kaggle/input/cat-dog-dataset/CatsDogs/Dogs'):
  if filename.endswith('.jpg'):
    try:
      img = Image.open('/kaggle/input/cat-dog-dataset/CatsDogs/Dogs/'+filename) # open the image file
      img.verify() # verify that it is, in fact an image
    except (IOError, SyntaxError) as e:
      print('Damaged Image:', filename)
```
